```python
from cosmiclib import Compressor, Optimizer, generate_sequence_plot
```

```python
compressor1 = Compressor('Compressor 1', qmin=8, qmax=40, wmin=26, wmax=56)
compressor2 = Compressor('Compressor 2', qmin=8, qmax=40, wmin=32, wmax=52)
compressor3 = Compressor('Compressor 3', qmin=13, qmax=127, wmin=77, wmax=149)
```

```python
compressors = [compressor1, compressor2, compressor3]
my_optimizer = Optimizer(compressors=compressors)
fig = generate_sequence_plot(my_optimizer)
fig.layout.template = 'plotly_dark'
fig.update_layout(paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)')
```

```python
fig.show()
```

```python
import plotly.graph_objs as go

fig = go.Figure(data=go.Scatter(x = [compressor1.qmin, compressor1.qmax], y = [compressor1.wmin, compressor1.wmax], mode='lines', name="Compressor 1"), layout_xaxis_range=[0,130], layout_yaxis_range=[0, 160])
fig.add_trace(go.Scatter(x = [compressor2.qmin, compressor2.qmax], y = [compressor2.wmin, compressor2.wmax], mode='lines', name="Compressor 2"))
fig.add_trace(go.Scatter(x = [compressor3.qmin, compressor3.qmax], y = [compressor3.wmin, compressor3.wmax], mode='lines', name="Compressor 3"))
fig.update_layout(title='Henderson 1 Low Stage Compressors', xaxis_title='q', yaxis_title='w', width=960, height=600, paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)')
fig.layout.template = 'plotly_dark'
fig.show()
```
